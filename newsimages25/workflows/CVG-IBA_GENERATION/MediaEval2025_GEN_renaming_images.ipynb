```python
# --- Renaming the images as per the convention of MediaEval 2025 NewsImage Gen Task
# Large Task

import os
import csv
import shutil

# === Paths ===
input_folder = r"D:\MediaEval 2025\Large Dataset results\Final" # modify according to your images source folder
output_folder = r"D:\MediaEval 2025\CVG-IBA\GEN_PromptForge_LARGE" # modify according to your images destination folder
csv_file = os.path.join(output_folder, "image_seed_mapping_large.csv") # this scv contains the seed number information for reproducibility

# Make sure output folder exists
os.makedirs(output_folder, exist_ok=True)

# Collect records
records = []

for filename in os.listdir(input_folder):
    if filename.lower().endswith(".png"):
        parts = filename.split("_")
        if len(parts) >= 2:
            image_num = parts[0]       # before first "_"
            seed_num = parts[1]        # after first "_"

            # New filename (Use the name as desired)
            new_name = f"{image_num}_CVG-IBA_PromptForge.png"

            # Copy to output folder
            src = os.path.join(input_folder, filename)
            dst = os.path.join(output_folder, new_name)
            shutil.copy2(src, dst)  # keeps metadata

            # Save mapping
            records.append([image_num, seed_num])

# Write CSV
with open(csv_file, "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["Image Number", "Seed Number"])
    writer.writerows(records)

print("✅ Renaming and copying complete.")
print(f"📂 Images saved to: {output_folder}")
print(f"📄 CSV saved to: {csv_file}")
```
